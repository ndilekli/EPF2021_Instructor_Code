```python
# Write a function that will take in the date range,
# (For example, day 0 to day 10 will be entered as [0, 10])
# and sum (add) up the min, max, mean, and std values
# across each day

# Add coefficients for min, max, mean and std
# Multiply 4 coefficients by 4 variables
# makeChart(data, [5,15], -1, 2, 0.5, 2)
```

```python
def makeChart(mydata, day_range):
    max_val =numpy.max(mydata[:,day_range[0]:day_range[1]], axis = 0)
    min_val = numpy.min(mydata[:,day_range[0]:day_range[1]], axis = 0)
    ave_val = numpy.mean(mydata[:,day_range[0]:day_range[1]], axis = 0) 
    std_val = numpy.std(mydata[:,day_range[0]:day_range[1]], axis = 0)
    result = max_val + min_val + ave_val + std_val
    sum_plot = matplotlib.pyplot.plot(result)
    matplotlib.pyplot.show()
    
makeChart(data, [0,40])
```

```python
# How the enter the correct X axis labels?

def makeChartCoeff(mydata, day_range, cmax, cmin, cave, cstd):
    max_val =numpy.max(mydata[:,day_range[0]:day_range[1]], axis = 0)
    min_val = numpy.min(mydata[:,day_range[0]:day_range[1]], axis = 0)
    ave_val = numpy.mean(mydata[:,day_range[0]:day_range[1]], axis = 0) 
    std_val = numpy.std(mydata[:,day_range[0]:day_range[1]], axis = 0)
    result = cmax * max_val + cmin * min_val + cave * ave_val + cstd * std_val
    
    x_labels = numpy.arange(day_range[0], day_range[1])
    sum_plot = matplotlib.pyplot.plot(x_labels, result)
    matplotlib.pyplot.show()
    
makeChartCoeff(data, [20,30], 1, 1.5, -2.4, 0.5)
```
